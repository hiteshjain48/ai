```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='most_frequent')
```

```python
gender = pd.DataFrame(acad['GENDER'])
```

```python
acad['GENDER'] = imputer.fit_transform(gender)
```

```python
placement = pd.DataFrame(acad['PLACEMENT'])
honor = pd.DataFrame(acad['HONOR_OPTED_OR_NOT'])
edu = pd.DataFrame(acad['EDUCATION_TYPE'])
prog = pd.DataFrame(acad['ACADEMIC_PROGRAM'])
```

```python
acad['PLACEMENT'] = imputer.fit_transform(placement)
acad['HONOR_OPTED_OR_NOT'] = imputer.fit_transform(honor)
acad['EDUCATION_TYPE'] = imputer.fit_transform(edu)
acad['ACADEMIC_PROGRAM'] = imputer.fit_transform(prog)
```

```python
acad['COURSE 1 MARKS'] = acad['COURSE 1 MARKS'].replace(np.NaN, 0)
acad['COURSE 2 MARKS'] = acad['COURSE 2 MARKS'].replace(np.NaN, 0)
acad['COURSE 3 MARKS'] = acad['COURSE 3 MARKS'].replace(np.NaN, 0)
acad['COURSE 4 MARKS'] = acad['COURSE 4 MARKS'].replace(np.NaN, 0)
acad['COURSE 5 MARKS'] = acad['COURSE 5 MARKS'].replace(np.NaN, 0)
```

```python
acad.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
acad['PLACEMENT'] = encoder.fit_transform(acad['PLACEMENT'])
```

```python
acad['GENDER'] = encoder.fit_transform(acad['GENDER'])
acad['HONOR_OPTED_OR_NOT'] = encoder.fit_transform(acad['HONOR_OPTED_OR_NOT'])
acad['EDUCATION_TYPE'] = encoder.fit_transform(acad['EDUCATION_TYPE'])
acad['ACADEMIC_PROGRAM'] = encoder.fit_transform(acad['ACADEMIC_PROGRAM'])
```

```python
acad.head()
```

```python
acad['OVEARLL_GRADE'] = encoder.fit_transform(acad['OVEARLL_GRADE'])
```
